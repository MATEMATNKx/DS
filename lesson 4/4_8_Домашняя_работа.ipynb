```python
from sklearn.datasets import load_boston
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
boston_dataset = load_boston()
X = boston_dataset.data
y = boston_dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
# -- ВАШ КОД ТУТ --




from sklearn.linear_model import SGDRegressor
from scipy.spatial import distance
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


sgd_regressor = SGDRegressor(learning_rate='constant', eta0=1e-8, fit_intercept=True, random_state=42)

#инициализируем веса, епсилон(окрестность)
w_current, epsilon = np.random.random(13), 1e-4
weight_evolution, mse_evolution = [], [] #изменение весов и ошибка на валидации
for step in list(range(8000)):
    sgd_regressor = sgd_regressor.partial_fit(X_train,y_train)
    weight_evolution.append(distance.euclidean(w_current, sgd_regressor.coef_))
    
    #если изменение весов находится в окрестности епсилона
    if weight_evolution[-1]<epsilon:
        print('weight_evolution is %d'% weight_evolution[-1])
        print("Итерация остановлена на шаге %d" % step); break
    
    mse_evolution.append(mean_squared_error(y_test, sgd_regressor.predict(X_test) )    )
    w_current = sgd_regressor.coef_.copy()
plt.plot(range(step), mse_evolution)
```

```python
from sklearn.datasets import load_boston
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

boston_dataset = load_boston()
X = boston_dataset.data
y = boston_dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
# -- ВАШ КОД ТУТ --

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


from sklearn.linear_model import SGDRegressor
from scipy.spatial import distance
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



sgd_regressor = SGDRegressor(learning_rate='constant', eta0=1e-5, fit_intercept=True, random_state=42)


#инициализируем веса, епсилон(окрестность)
w_current, epsilon = np.random.random(13), 1e-4
weight_evolution, r2_evolution = [], [] #изменение весов и ошибка на валидации
for step in list(range(8000)):
    sgd_regressor = sgd_regressor.partial_fit(X_train,y_train)
    weight_evolution.append(distance.euclidean(w_current, sgd_regressor.coef_))
    
    #если изменение весов находится в окрестности епсилона
    if weight_evolution[-1]<epsilon:
        print('weight_evolution is %d'% weight_evolution[-1])
        print("Итерация остановлена на шаге %d" % step); break
    
    r2_evolution.append(r2_score(y_test, sgd_regressor.predict(X_test) )    )
    w_current = sgd_regressor.coef_.copy()
plt.plot(range(0, len(r2_evolution)), r2_evolution)
plt.title(str(r2_evolution[-1]))
plt.show()
```
